```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv('../data/raw/diabetes_prediction_dataset.csv')
```

```python
assets_dir = "../assets/"
if not os.path.exists(assets_dir):
    os.makedirs(assets_dir)
```

```python
ax = sns.countplot(x='gender', hue='diabetes', data=df, palette= "Set1")
for label in ax.containers:
    ax.bar_label(label)
plt.savefig("../assets/countplot1.png")
plt.show()
```

```python
num_features = ['gender','age','hypertension','heart_disease','smoking_history','bmi','HbA1c_level','blood_glucose_level','diabetes']
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust the layout
fig, axs = plt.subplots(5, 2, figsize=(15, 20))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

# List of features
features = ['gender', 'age', 'hypertension', 'heart_disease', 'smoking_history', 'bmi', 'HbA1c_level', 'blood_glucose_level', 'diabetes']

# Plot each feature in a separate subplot
for i, feature in enumerate(features):
    row = i // 2
    col = i % 2
    sns.boxplot(data=df, x=feature, ax=axs[row, col])

# Remove the extra subplot (the last one in the grid)
fig.delaxes(axs[4, 1])
plt.savefig("../assets/boxplot8.png")


plt.show()
```
